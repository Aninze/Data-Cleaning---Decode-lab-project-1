```python
# Import neccessary important python libs:

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# Import Dataset for Data analysis 

df = pd.read_csv(r"C:\Users\aninz\Desktop\Class materials\Dataset for Data Analytics.csv")
```

```python
# view df table

df.head()
```

```python
# checking the table dimension

print('Number of Columns: ',df.shape[1])
print('Number of rows: ',df.shape[0])
```

```python
# Checking for missing values 

df.isnull().sum()
```

<b>Note!</b> Coupon Code column has 309 missing values

```python
# Determine the most frequent value in the CouponCode

df_CouponCode_mode = df.CouponCode.mode()[0]
print('Mode value: ',df_CouponCode_mode)

# Replace the missing value with the mode value

df.CouponCode.fillna(df_CouponCode_mode, inplace = True)

# Re-check the missing value
print('Number of missing value: ',df.CouponCode.isnull().sum())
```

```python
# Determine the unique values in the CouponCode 

print('Unique Value: ',df.CouponCode.unique())

# Remove "10" and "15"  from "SAVE10" and "WINTER15" respectively

df.CouponCode = df.CouponCode.str.replace(r'\d+','',regex = True)
```

```python
print('Unique Value: ',df.CouponCode.unique())
```

```python
# Check missing value

print( df.columns[df.isnull().sum() >1].to_list() )
```

```python
# Export as cleaned dataset

df.to_csv('Dataset for data analysis_clean.csv',index = False)
```
